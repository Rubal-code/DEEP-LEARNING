```python
import pandas as pd
import gensim
```

```python
df=pd.read_json("/kaggle/input/datasets/hakim11/sports-and-outdoors-5/Sports_and_Outdoors_5.json",lines=True
              )
```

```python
review_text=df.reviewText.apply(gensim.utils.simple_preprocess)
```

```python
review_text.loc[0]
```

```python
model=gensim.models.Word2Vec(
    window=10,
    min_count=5,
    workers=4
)
```

```python
model.build_vocab(review_text,progress_per=1000)
```

```python
model.train(review_text,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar("already")
```

```python
model.wv.similarity(w1="already",w2="danger")
```
