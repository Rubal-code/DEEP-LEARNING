```python
import tensorflow_hub as hub
import tensorflow as tf
```

```python
preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

taking input

```python
text_input=tf.keras.layers.Input(shape=(), dtype=tf.string,name='text')
```

preprocessing

```python
preprocess=hub.KerasLayer(preprocess_url)
encoder_input=preprocess(text_input)
```

bert brain

```python
bert_encoder=hub.KerasLayer(encoder_url,trainable=True)
outputs=bert_encoder(encoder_input)
```

final decision layer

Converts meaning → result

Output ≈ 1 → Positive
Output ≈ 0 → Negative

```python
pooled_output=outputs['pooled_output']
output=tf.keras.layers.Dense(1,activation='sigmoid',)(pooled_output)
```

build the model

```python
model = tf.keras.Model(inputs=text_input, outputs=output)
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```
