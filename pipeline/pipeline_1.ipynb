```python
import tensorflow as tf
```

```python
daily_sales_numbers=[21,22,-108,31,32,-109,33,34,-110,35,36,-111,37,38,-112,39,40,-113]
```

```python
tf_dataset=tf.data.Dataset.from_tensor_slices(daily_sales_numbers) # Create a TensorFlow dataset from the daily sales numbers which means we are converting the list of daily sales numbers into a format that TensorFlow can work with for further processing and analysis.
tf_dataset
```

```python
for sales in tf_dataset.as_numpy_iterator():# Iterate through the TensorFlow dataset using as_numpy_iterator() to convert each element of the dataset into a NumPy array format, which allows us to easily access and manipulate the individual sales numbers.
    print(sales)
```

```python
tf_dataset=tf_dataset.filter(lambda x:x>0) # Filter the dataset to include only positive sales numbers, which means we are removing any negative sales numbers from the dataset, leaving us with only the valid sales data for further analysis.
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
tf_dataset=tf_dataset.map(lambda x: x*10) # Map the dataset to multiply each sales number by 10, which means we are scaling the sales numbers by a factor of 10, potentially for normalization or to adjust the scale of the data for further analysis or modeling.
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
tf_dataset=tf_dataset.shuffle(3) # Shuffle the dataset with a buffer size of 3, which means we are randomly shuffling the order of the sales numbers in the dataset using a buffer of 3 elements, helping to ensure that the data is not in a specific order and can improve the performance of machine learning models trained on this data.
for sales in tf_dataset.as_numpy_iterator():    
    print(sales)
```

```python
for sales_batch in tf_dataset.batch(3): # Batch the dataset with a batch size of 3, which means we are grouping the sales numbers into batches of 3 for more efficient processing and training of machine learning models, allowing us to work with multiple data points at once.
    print(sales_batch)
```

##### alll these steps in the single line


```python
tf_dataset=tf.data.Dataset.from_tensor_slices(daily_sales_numbers)
tf_dataset=tf_dataset.filter(lambda x:x>0).map(lambda y: y*10).shuffle(3).batch(3) # Chain multiple transformations together to create a more complex data processing pipeline, which means we are applying multiple operations (filtering, mapping, shuffling, and batching) in a single line of code to efficiently process the sales data for machine learning tasks.
for sales_batch in tf_dataset.as_numpy_iterator():
    print(sales_batch)
```

```python
train_size=int(image_count*0.8)
train_ds=images_ds.take(train_size)
test_ds=images_ds.skip(train_size)
```

```python
import os
def get_label(file_path):
    # split the path into the segments
    parts=tf.strings.split(file_path,os.path.sep)
    # grab the parent folder name
    return parts[-2]
```

```python
get_label("images\\dog\\20 Reasons Why Cats Make the Best Pets....jpg")
```

```python
def process_image(file_path):
    label = get_label(file_path)
    img = tf.io.read_file(file_path) # load the raw data from the file as a string
    img = tf.image.decode_jpeg(img)
    img = tf.image.resize(img, [128, 128])
    return img, label
```

```python
img, label = process_image("images\\cat\\20 Reasons Why Cats Make the Best Pets....jpg")
img.numpy()[:2]
```

```python
train_ds = train_ds.map(process_image)
test_ds = test_ds.map(process_image)
```

```python
for image, label in train_ds.take(1):
    print("****",image)
    print("****",label)
```

```python
def scale(image, label):
    return image/255, label
```

```python
train_ds = train_ds.map(scale)
```
