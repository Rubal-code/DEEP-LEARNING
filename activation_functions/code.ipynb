# implementation of the activation functions in the python

***Sigmoid***

```python
import math

def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(10)
```

***tanh***

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(100)
```

***ReLU***

```python
def ReLU(x):
    return max(0,x)
```

```python
ReLU(10)
```

***Leaky ReLU***

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-10)
```
